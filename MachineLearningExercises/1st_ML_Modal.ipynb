Import the Libraries

```
import pandas as pd
import numpy as np
%matplotlib inline
```

Selecting Data for Modeling

```
from sklearn.datasets import load_iris
data = load_iris()
```

```
data['target_names']
```

Data Prep From Modeling

Selecting Features & Target ('Meat' & 'Desert') for Modeling

```
x = data['data']
y = data['target']
```

MODELING STEP # 1 -- Define the Modal

```
from sklearn.tree import DecisionTreeClassifier
modal = DecisionTreeClassifier()
```

MODELING STEP # 2 -- Fit the Modal

```
modal.fit(x, y)
```

MODELING STEP # 3 -- Run Prediction 

```
new_data = pd.DataFrame(x)
predict = modal.predict(new_data.head(5))
```

MODELING STEP # 4 -- Evaluate the Modal

```
print(predict)
```
