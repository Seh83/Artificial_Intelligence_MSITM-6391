Import the Libraries

```
import pandas as pd
import numpy as np
%matplotlib inline
```

Selecting Data for Modeling

```
data = pd.read_csv('/content/drive/My Drive/MISTM/2/data/melb_data.csv')
```

Data Prep From Modeling

```
#Drop null and empty values from the dataset
data = data.dropna(axis=0)
data.head(5)
```

```
data.keys()
```

Selecting Features & Target ('Meat' & 'Desert') for Modeling

```
#Create Featurte
x = data[['Rooms', 'Distance', 'Postcode', 'Bedroom2', 'Bathroom', 'Car',
       'Landsize', 'BuildingArea', 'YearBuilt',  'Lattitude',
       'Longtitude', 'Propertycount']]
#Create Target 
y = data['Price']
```

MODELING STEP # 1 -- Define the Modal

```
from sklearn.tree import DecisionTreeRegressor

modal = DecisionTreeRegressor(random_state=1)
```

MODELING STEP # 2 -- Fit the Modal

```
modal.fit(x, y)
```

MODELING STEP # 3 -- Run Prediction 

```
new_data = x.head(3)
predict = modal.predict(new_data)
```

MODELING STEP # 4 -- Evaluate the Modal

```
print(predict)
```
